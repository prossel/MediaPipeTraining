# Split a dataset into training and validation sets

Export the dataset from Label Studio in COCO format, unzip it and rename the folder to `flatdataset`.

Then run the following cell to split the dataset into training and validation sets:


````python
# Shift category id's and add background class with id 0
import json

fpath = "flatdataset/result.json"

# load coco json
with open(fpath, "r", encoding="utf-8") as f:
    coco = json.load(f) 
    
# shift category id's by 1
for cat in coco["categories"]:
    cat["id"] += 1

# add background class
coco["categories"].insert(0, {"id": 0, "name": "background"})   

# shift category id's in annotations
for ann in coco["annotations"]:
    ann["category_id"] += 1
        
# save coco json to a new file with the _bg suffix
with open(fpath.replace(".json", "_bg.json"), "w", encoding="utf-8") as f:
    json.dump(coco, f, ensure_ascii=False, indent=2)    
print("Category ids shifted and background class added.")

````
